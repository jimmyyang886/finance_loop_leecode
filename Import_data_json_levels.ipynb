```python
import pandas as pd
import numpy as np
import json

data=json.load(open('2017Q3/1101-2017-Q3.json'))['data']
#print(data[0].get('Balance'))

# for _ in data[0].get('Balance'):
#     print(_[0][0], _[0][1], _[1])

code=data[0].get('Balance')

for i in range(len(code)):
    print(code[i][0], code[i][1])
```

```python
lv1_dict={}
# lv2_dict={}
# lv3_dict={}


for i in range(len(code)):
    if code[i][0][0]==1:
        lv1=code[i][0][1]
        lv2_dict={}
        
    elif code[i][0][0]==2:
        lv2=code[i][0][1]
        lv3_dict={}
        
        if code[i][1] == '':
            #lv2_list.append({code[i][0][1]:[]})
            lv2_dict[code[i][0][1]]=lv3_dict
        else:
             #lv2_list.append({code[i][0][1]:code[i][1]})
            lv2_dict[code[i][0][1]]=code[i][1]
            
        lv1_dict[lv1]=lv2_dict
        

    elif code[i][0][0]==3:  
        lv3=code[i][0][1]
        lv4_dict={}

        lv2_dict=lv1_dict.get(lv1)

        if code[i][1] == '':
            lv3_dict[code[i][0][1]]=lv4_dict
            
        else:
            lv3_dict[code[i][0][1]]=code[i][1]
        lv2_dict[lv2]=lv3_dict
        
        
    elif code[i][0][0]==4:

        lv4=code[i][0][1]
        lv5_dict={}

        if code[i][1] == '':
            lv4_dict[code[i][0][1]]=lv5_dict
        else:
            lv4_dict[code[i][0][1]]=code[i][1]
              
        #lv3_dict=lv2_dict[lv2][-1]
        lv3_dict=lv2_dict.get(lv2)
        lv3_dict[lv3]=lv4_dict 


    elif code[i][0][0]==5: 
        
        lv5=code[i][0][1]
        lv6_dict={}
        
        if code[i][1] == '':
            lv5_dict[code[i][0][1]]=lv6_dict
        else:
            lv5_dict[code[i][0][1]]=code[i][1]
        lv4_dict=lv3_dict.get(lv3)
        lv4_dict[lv3]=lv5_dict  

    
    elif code[i][0][0]==6:
        
        lv6=code[i][0][1]
        lv7_dict={}
        if code[i][1] == '':
            lv6_list.append({code[i][0][1]:[]})
        else:
            lv6_dict[code[i][0][1]]=code[i][1]
        lv5_dict=lv4_dict.get(lv3)
        lv5_dict[lv4]=lv6_dict   

print(lv1_dict)   
# with open('2017Q3/1101-2017-Q3_test.json', 'w') as f:
#         json.dump(lv1_dict, f)  
```

```python
df=pd.json_normalize(lv1_dict, max_level=7)
df
```
